```python
pairs = [
    ('SPY', 'GLD'),
    ('SPY', 'TLT'),
    ('SPY', 'UUP'),
    ('GLD', 'TLT'),
    ('GLD', 'UUP'),
    ('TLT', 'UUP')
]

fig, axes = plt.subplots(3, 2, figsize=(16, 12))
axes = axes.flatten()

for i, (a, b) in enumerate(pairs):
    rc = returns[a].rolling(window=60).corr(returns[b])
    mean = rc.mean()
    std = rc.std()
    upper = mean + 2 * std
    lower = mean - 2 * std
    
    axes[i].plot(rc, color='blue', linewidth=0.8)
    axes[i].axhline(upper, color='red', linestyle='--', linewidth=0.8)
    axes[i].axhline(lower, color='red', linestyle='--', linewidth=0.8)
    axes[i].axhline(mean, color='green', linestyle='--', linewidth=0.8)
    axes[i].fill_between(rc.index, lower, upper, alpha=0.1, color='green')
    axes[i].set_title(f'{a} vs {b}')
    axes[i].set_ylabel('Correlation')

plt.suptitle('Cross-Asset Correlation Anomaly Detector', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
print("=" * 60)
print("ANOMALY DETECTION SUMMARY")
print("=" * 60)

for a, b in pairs:
    rc = returns[a].rolling(window=60).corr(returns[b])
    mean = rc.mean()
    std = rc.std()
    upper = mean + 2 * std
    lower = mean - 2 * std
    
    anomalies = rc[(rc > upper) | (rc < lower)]
    
    print(f"\n{a} vs {b}:")
    print(f"  Historical Mean: {mean:.3f}")
    print(f"  Upper Threshold: {upper:.3f}")
    print(f"  Lower Threshold: {lower:.3f}")
    print(f"  Number of Anomaly Days: {len(anomalies)}")
    
    if len(anomalies) > 0:
        print(f"  First Anomaly: {anomalies.index[0].date()}")
        print(f"  Last Anomaly:  {anomalies.index[-1].date()}")

print("\n" + "=" * 60)
```
